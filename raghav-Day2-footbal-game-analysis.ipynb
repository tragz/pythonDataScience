# Random walks

```python
import random
position = 0
walk = [ position]
steps = 1000

for i in range(steps):
    step = 1 if random.randint(0,1) else -1  # logic of random walk if number is (0,1) add 1 else subtract 1
    position += step
    walk.append( position)
 

#plt.plot(walk)
plt.plot(walk[:200]) # plot first 100 steps
plt.plot(walk[201:400])
plt.plot(walk[401:600])
plt.plot(walk[601:800])
plt.plot(walk[801:1000])

plt.show()
```
